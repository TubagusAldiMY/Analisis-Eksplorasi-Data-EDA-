```python
df = df.drop(columns=['sku','url','sale_price/amount_with_symbol',
                 'retail_price/amount_with_symbol','category_id','description'])
df.head()
```

```python
df = df.rename(columns={'product_id':'Product_ID','title':'Title','color':'Color',
                  'sale_price/amount':'Selling_Price','retail_price/amount':'Retail_Price',
                  'discount_percentage':'Discount_Percentage','category_name':'Category_Name',
                  'reviews_count':'Reviews_Count','average_rating':'Average_Rating'})
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df['Selling_Price'] = df['Selling_Price'].apply(lambda x: int(x))
df['Retail_Price'] = df['Retail_Price'].apply(lambda x: int(x))
df['Average_Rating'] = df['Average_Rating'].apply(lambda x: float(x))
```

```python
df['Discount_Percentage'] = df['Discount_Percentage'].astype(float)
df['Discount_Percentage']
```

```python
df.dtypes
```

```python
high_sellingprice = df.groupby('Title')['Selling_Price'].sum().sort_values(ascending=False).head(5)
```

```python
ax = high_sellingprice.plot(kind='bar', legend=None, color='purple')

ax.set_xlabel('Product Title')
ax.set_ylabel('Sum of Selling Price')
ax.set_title('Top 5 Highest Selling Price Products')

plt.show()
```

```python
high_sellingprice_category = df.groupby('Category_Name')['Selling_Price'].sum().sort_values(ascending=False).head(5)


high_sellingprice_category
```

```python
ax = high_sellingprice_category.plot(kind='barh', color='green', legend=None)

ax.set_xlabel('Sum of Selling_Price')
ax.set_title('Top 5 Category by Selling_Price')

plt.show()
```

```python
color_high_retail = df.groupby('Color')['Retail_Price'].sum() \
                    .to_frame('Sum').sort_values(by='Sum', ascending=False) \
                    .head(5)
color_high_retail
```
